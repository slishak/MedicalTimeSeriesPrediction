```python
import torch

from biophysical_models import models

cvs = models.SmithCardioVascularSystem()

v_lv = torch.linspace(0.02, 0.15, 100)
v_rv = torch.linspace(0.02, 0.15, 100)
e_t = torch.linspace(0, 1, 10)
V_lv, V_rv, E_t = torch.meshgrid(v_lv, v_rv, e_t, indexing='ij')
```

```python
V_spt = cvs.solve_v_spt(V_lv, V_rv, E_t)
```

```python
V_spt.shape
```

```python
import matplotlib.pyplot as plt

for i in range(10):
    plt.figure()
    plt.contourf(V_lv[:, :, i], V_rv[:, :, i], V_spt[:, :, i])
    plt.colorbar()
```
